## Key Findings

- The dataset contains 284,807 rows and 31 columns, indicating a large dataset suitable for training machine learning models.

- The target column is `Class`, where 0 represents normal transactions and 1 represents fraudulent transactions.

- The dataset is highly imbalanced, with fraud cases representing approximately 0.17% of the dataset. 
- This  imbalance will impact model performance, making accuracy an unreliable metric. Specialized techniques such as class weighting, resampling, or anomaly detection will be required.

- The features `V1` to `V28` are PCA-transformed numerical features. This reduces the need for feature engineering.

- The `Amount` and `Time` columns are not PCA-transformed and will require scaling to ensure they are on a similar scale to the other features.

- No missing values were found in the dataset,no imputation is required.

- A total of 1081 duplicate rows were identified. These duplicates may introduce bias into the model if not handled they will be addressed during preprocessing.

- Due to the nature of fraud detection, recall will be an important evaluation metric, as failing to detect fraudulent transactions can have significant real-world consequences.

- The presence of duplicate transactions raises an important consideration: whether these represent true repeated transactions or redundant data. This decision will be evaluated during preprocessing.
